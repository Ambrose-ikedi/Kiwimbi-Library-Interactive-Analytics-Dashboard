**Kiwimbi Library Interactive Analytics Dashboard**

```python
import pandas as pd
```

```python
url = '/content/Kiwimbi Kenya Data Submission (Responses) (2).xlsx'
```

```python
df = pd.read_excel(url)
```

```python
df.head()
```

```python
import plotly.express as px
```

```python
# Melt the DataFrame to transform monthly columns into rows, keeping 'NAMES:' as an identifier
df_melted = df.melt(id_vars=["NAMES:"], var_name="Month", value_name="Status")

# Count the occurrences of 'Yes' and 'No' for each name
df_counts = df_melted.groupby(["NAMES:", "Status"]).size().reset_index(name="Count")

# Create the bar chart using the transformed data
fig1 = px.bar(
    df_counts,
    x="Status", # Now 'Status' is a column with 'Yes' or 'No'
    y="Count",  # Now 'Count' is a column with the number of occurrences
    color="Status", # Differentiate 'Yes' and 'No' bars by color
    facet_col="NAMES:", # Create a subplot for each person
    title="Count of Yes/No Responses Per Person Across All Months"
)
```

```python
# Calculate counts of 'Yes' and 'No' responses per month
df_monthly_counts = df_melted.groupby(['Month', 'Status']).size().reset_index(name='Count')

fig2 = px.line(
    df_monthly_counts,
    x="Month",
    y="Count",
    color="Status", # Differentiate lines by 'Yes' or 'No'
    title="Trend of Yes/No Responses Over Time"
)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

dashboard = make_subplots(
    rows=2, cols=1,
    subplot_titles=("Yes/No Responses Per Person", "Trend of Yes/No Responses Over Time")
)

# Add traces from fig1 (bar chart) to the first row of the dashboard
for trace in fig1['data']:
    dashboard.add_trace(trace, row=1, col=1)

# Add traces from fig2 (line chart) to the second row of the dashboard
for trace in fig2['data']:
    dashboard.add_trace(trace, row=2, col=1)

dashboard.update_layout(height=800, title="Analytics Dashboard", showlegend=True)
```

```python
dashboard.write_html(
    "dashboard.html",
    full_html=True,
    include_plotlyjs="cdn"
)
```
